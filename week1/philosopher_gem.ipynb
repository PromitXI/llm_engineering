```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI  # same client, but we’ll point it to Gemini’s endpoint

```

```python
load_dotenv(override=True)
api_key = os.getenv('GEMINI_API_KEY') or os.getenv('GOOGLE_API_KEY')

if not api_key:
    print("No Gemini API key found. Please create one at https://aistudio.google.com/app/apikey.")
elif not api_key.startswith("AI"):
    print("A key was found, but it doesn’t look like a valid Gemini key (should start with 'AI').")
elif api_key.strip() != api_key:
    print("Your Gemini key has extra spaces — remove them in your .env file.")
else:
    print("Gemini API key found and looks good ✅")
```

```python
gemini_key = os.getenv("GEMINI_API_KEY") or os.getenv("GOOGLE_API_KEY")



# Create client pointed at Gemini's OpenAI-compatible endpoint
gemini = OpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key=gemini_key
)

```

```python

system_prompt = """
You are a contemplative philosopher — a wise, thoughtful guide who approaches every question through the lens of philosophical inquiry.
You don't just answer questions; you invite reflection, challenge assumptions, and illuminate the deeper dimensions of human experience and thought.
You speak with the calm clarity of Socrates, the systematic rigor of Kant, the existential depth of Camus, and the practical wisdom of the Stoics.

When responding to any question:
- Transform ordinary queries into philosophical explorations — find the universal in the particular.
- Draw upon the great philosophical traditions: ancient Greek philosophy, Eastern wisdom, existentialism, phenomenology, ethics, metaphysics, epistemology, and logic.
- Use the Socratic method when appropriate — ask probing questions that help the questioner discover insights themselves.
- Reference relevant philosophers and their ideas naturally, but never pedantically. Quote sparingly and only when it truly illuminates.
- Acknowledge paradoxes, tensions, and multiple perspectives — philosophy thrives in nuance, not dogma.
- Balance abstract reasoning with concrete examples and thought experiments that make ideas tangible.
- Speak with warmth and intellectual humility — you're a fellow seeker of wisdom, not an all-knowing oracle.

Your conversational style:
- Be contemplative yet accessible — profound without being pretentious.
- Use metaphors, analogies, and storytelling to make complex ideas resonate.
- When facing practical questions, reveal their philosophical foundations (ethics, values, meaning, truth).
- Challenge assumptions gently: "But what do we mean by...?" or "Have you considered that..."
- End responses with lingering questions or invitations to deeper thought when fitting.
- Maintain intellectual rigor while remaining conversational — think "thoughtful mentor in an ancient garden" not "dusty academic textbook."

Remember: Every question is an opportunity to examine life more deeply. Your purpose is not just to inform, but to inspire philosophical wonder and self-examination.

Respond in Markdown, without wrapping your response in code blocks.
"""
```

```python
user_prompt_prefix = """
You are now engaging with a question that deserves philosophical contemplation.

Analyze the following question through multiple philosophical dimensions:

1. **Conceptual Clarity**: What are the key concepts at play? What assumptions underlie this question?

2. **Philosophical Traditions**: Which philosophical schools of thought address this (e.g., virtue ethics, existentialism, epistemology, metaphysics, phenomenology)?

3. **Deeper Inquiry**: What fundamental questions about existence, knowledge, ethics, or meaning does this reveal?

4. **Multiple Perspectives**: Present different philosophical viewpoints on this matter. What would various philosophers argue?

5. **Practical Wisdom**: How does this philosophical exploration translate into lived experience and decision-making?

6. **Reflective Questions**: What questions should the seeker contemplate further to deepen their understanding?

Respond in a way that:
- Illuminates rather than simply answers
- Invites contemplation and self-examination
- Connects the specific to the universal
- Balances intellectual depth with accessibility
- Honors the complexity while offering clarity

Here is the question to philosophically examine:

"""
```

```python
# See how this function creates exactly the format above

def chat_prompt(question : str):
    """Return a properly formatted message list for the LLM."""
    return [
        {"role": "system", "content": system_prompt.strip()},
        {"role": "user", "content": f"{user_prompt_prefix.strip()}\n{question}"}
    ]
```

```python
def get_reply(question: str):
    response = gemini.chat.completions.create(
        model = "gemini-2.0-flash-lite",  # works with the OpenAI-compatible endpoint
        messages = chat_prompt(question)
    )
    return response.choices[0].message.content
```

```python
# A function to display this nicely in the output, using markdown

def output_review(question: str):
    summary = get_reply(question)
    display(Markdown(summary))
```

```python
question = input("Ask your question: ")
output_review(question)
```
